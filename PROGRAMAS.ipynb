```python
class Persona:
    def __init__(self, nombre, gmail):
        self.nombre = nombre
        self.gmail = gmail

class Usuario(Persona):
    def __init__(self, nombre, gmail):
        super().__init__(nombre, gmail)
        self.reservas = []

    def hacer_reserva(self, reserva):
        self.reservas.append(reserva)

class Empleado(Persona):
    def __init__(self, nombre, gmail, rol):
        super().__init__(nombre, gmail)
        self.rol = rol

    def agregar_funcion(self, cine, funcion):
        cine.funciones.append(funcion)

    def agregar_pelicula(self, cine, pelicula):
        cine.peliculas.append(pelicula)

    def agregar_promocion(self, cine, promocion):
        cine.promociones.append(promocion)

class Espacio:
    def __init__(self, nombre):
        self.nombre = nombre

class Sala(Espacio):
    def __init__(self, nombre, tipo, capacidad):
        super().__init__(nombre)
        self.tipo = tipo
        self.capacidad = capacidad
        self.asientos_ocupados = set()

    def verificar_disponibilidad(self, asientos):
        return all(asiento not in self.asientos_ocupados for asiento in asientos)

    def reservar_asientos(self, asientos):
        if self.verificar_disponibilidad(asientos):
            self.asientos_ocupados.update(asientos)
            return True
        return False

class ZonaComida(Espacio):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.productos = {}

    def agregar_producto(self, producto, precio, cantidad):
        self.productos[producto] = {'precio': precio, 'cantidad': cantidad}

    def comprar_producto(self, producto, cantidad):
        if producto in self.productos and self.productos[producto]['cantidad'] >= cantidad:
            self.productos[producto]['cantidad'] -= cantidad
            return f"Compra exitosa: {cantidad} de {producto}"
        return "Producto no disponible o cantidad insuficiente"

class Pelicula:
    def __init__(self, titulo, duracion, clasificacion, genero):
        self.titulo = titulo
        self.duracion = duracion
        self.clasificacion = clasificacion
        self.genero = genero

class Promocion:
    def __init__(self, descripcion, descuento):
        self.descripcion = descripcion
        self.descuento = descuento

class Funcion:
    def __init__(self, pelicula, sala, horario):
        self.pelicula = pelicula
        self.sala = sala
        self.horario = horario

class Reserva:
    def __init__(self, usuario, funcion, asientos, promocion=None):
        self.usuario = usuario
        self.funcion = funcion
        self.asientos = asientos
        self.promocion = promocion
        if self.funcion.sala.reservar_asientos(asientos):
            usuario.hacer_reserva(self)
        else:
            raise ValueError("Asientos ya ocupados")

class Cine:
    def __init__(self, nombre):
        self.nombre = nombre
        self.funciones = []
        self.peliculas = []
        self.promociones = []

cinepolis = Cine("Cinepolis")

empleado1 = Empleado("Carlos Maximiliano", "carlos@gmail.com", "Conserje")
empleado2 = Empleado("Gustambo Cerati", "gusgus@gmail.com", "Administrador")
empleado3 = Empleado("Charly Garcia", "charlytos@gmail.com", "Cinepolito")

sala1 = Sala("Sala 11", "IMAX", 140)
sala2 = Sala("Sala 7", "3D", 100)
sala3 = Sala("Sala 9", "2D", 80)

pelicula1 = Pelicula("Inception", 148, "PG-13", "Ciencia Ficci√≥n")
pelicula2 = Pelicula("Avatar", 162, "PG-13", "Aventura")
pelicula3 = Pelicula("Scott Pilgrim vs. The World", 172, "PG-13", "Comedia")

funcion1 = Funcion(pelicula1, sala1, "18:20")
funcion2 = Funcion(pelicula2, sala2, "20:30")
funcion3 = Funcion(pelicula3, sala3, "18:30")

empleado1.agregar_funcion(cinepolis, funcion1)
empleado2.agregar_funcion(cinepolis, funcion2)
empleado3.agregar_funcion(cinepolis, funcion3)

usuario1 = Usuario("Juan Jorobas", "juanjo@gmail.com")
usuario2 = Usuario("Maria Lopez", "marifer@gmail.com")
usuario3 = Usuario("Diego Mora", "edimora@gmail.com")

reserva1 = Reserva(usuario1, funcion1, [1, 2, 3])
reserva2 = Reserva(usuario2, funcion2, [5, 6])
reserva3 = Reserva(usuario3, funcion3, [7, 8])

zona_comida = ZonaComida("Snack Bar")
zona_comida.agregar_producto("Popcorn", 5.0, 50)
zona_comida.agregar_producto("Soda", 3.0, 30)
zona_comida.agregar_producto("Nachos", 6.5, 20)

print(zona_comida.comprar_producto("Popcorn", 2))
print(zona_comida.comprar_producto("Soda", 35))

print(f"Reservas de {usuario1.nombre}: {[r.funcion.pelicula.titulo for r in usuario1.reservas]}")
print(f"Reservas de {usuario2.nombre}: {[r.funcion.pelicula.titulo for r in usuario2.reservas]}")
print(f"Reservas de {usuario3.nombre}: {[r.funcion.pelicula.titulo for r in usuario3.reservas]}")
```

```python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Cliente(Persona):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.historial_pedidos = []

    def realizar_pedido(self, pedido):
        self.historial_pedidos.append(pedido)

class Empleado(Persona):
    def __init__(self, nombre, rol):
        super().__init__(nombre)
        self.rol = rol

class ProductoBase:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class Bebida(ProductoBase):
    def __init__(self, nombre, precio, tamano, tipo, opciones_personalizadas=None):
        super().__init__(nombre, precio)
        self.tamano = tamano
        self.tipo = tipo
        self.opciones_personalizadas = opciones_personalizadas if opciones_personalizadas else {}

class Postre(ProductoBase):
    def __init__(self, nombre, precio, es_vegano=False, sin_gluten=False):
        super().__init__(nombre, precio)
        self.es_vegano = es_vegano
        self.sin_gluten = sin_gluten

class Inventario:
    def __init__(self):
        self.ingredientes = {}

    def agregar_ingrediente(self, nombre, cantidad):
        self.ingredientes[nombre] = self.ingredientes.get(nombre, 0) + cantidad

    def verificar_disponibilidad(self, ingredientes_requeridos):
        return all(self.ingredientes.get(ing, 0) >= cantidad for ing, cantidad in ingredientes_requeridos.items() if isinstance(cantidad, int))

    def descontar_ingredientes(self, ingredientes_requeridos):
        if self.verificar_disponibilidad(ingredientes_requeridos):
            for ing, cantidad in ingredientes_requeridos.items():
                if isinstance(cantidad, int):
                    self.ingredientes[ing] -= cantidad
            return True
        return False

class Pedido:
    def __init__(self, cliente):
        self.cliente = cliente
        self.productos = []
        self.estado = "pendiente"

    def agregar_producto(self, producto):
        self.productos.append(producto)

    def calcular_total(self):
        return sum(p.precio for p in self.productos)

class Promocion:
    def __init__(self, descuento, criterio):
        self.descuento = descuento
        self.criterio = criterio

    def aplicar_descuento(self, pedido):
        if self.criterio(pedido):
            return pedido.calcular_total() * (1 - self.descuento)
        return pedido.calcular_total()

inventario = Inventario()
inventario.agregar_ingrediente("leche de almendra", 2)
inventario.agregar_ingrediente("az√∫car", 5)
inventario.agregar_ingrediente("caf√©", 10)
inventario.agregar_ingrediente("harina", 10)

cliente1 = Cliente("Carlos")
cliente2 = Cliente("Ana")
cliente3 = Cliente("Luis")

bebida1 = Bebida("Caf√© Americano", 3.0, "Grande", "Caliente", {"caf√©": 1, "leche de almendra": 1})
bebida2 = Bebida("T√© Chai", 4.0, "Mediano", "Caliente", {"leche de almendra": 1})
bebida3 = Bebida("Latte", 5.0, "Grande", "Caliente", {"caf√©": 2, "leche de almendra": 3})
postre = Postre("Brownie", 2.5, False, False)

pedido1 = Pedido(cliente1)
pedido2 = Pedido(cliente2)
pedido3 = Pedido(cliente3)

if inventario.verificar_disponibilidad(bebida1.opciones_personalizadas):
    inventario.descontar_ingredientes(bebida1.opciones_personalizadas)
    pedido1.agregar_producto(bebida1)
    cliente1.realizar_pedido(pedido1)
    print(f"Pedido de {cliente1.nombre}: {pedido1.calcular_total()}$")
else:
    print(f"No hay suficientes ingredientes para preparar el pedido de {cliente1.nombre}.")

if inventario.verificar_disponibilidad(bebida2.opciones_personalizadas):
    inventario.descontar_ingredientes(bebida2.opciones_personalizadas)
    pedido2.agregar_producto(bebida2)
    pedido2.agregar_producto(postre)
    cliente2.realizar_pedido(pedido2)
    print(f"Pedido de {cliente2.nombre}: {pedido2.calcular_total()}$")
else:
    print(f"No hay suficientes ingredientes para preparar el pedido de {cliente2.nombre}.")

if inventario.verificar_disponibilidad(bebida3.opciones_personalizadas):
    inventario.descontar_ingredientes(bebida3.opciones_personalizadas)
    pedido3.agregar_producto(bebida3)
    cliente3.realizar_pedido(pedido3)
    print(f"Pedido de {cliente3.nombre}: {pedido3.calcular_total()}$")
else:
    print(f"No hay suficientes ingredientes para preparar el pedido de {cliente3.nombre}.")
```

```python
from datetime import datetime, timedelta

class Material:
    def __init__(self, titulo, estado='disponible'):
        self.titulo = titulo
        self.estado = estado

class Libro(Material):
    def __init__(self, titulo, autor, genero, estado='disponible'):
        super().__init__(titulo, estado)
        self.autor = autor
        self.genero = genero

class Revista(Material):
    def __init__(self, titulo, edicion, periodicidad, estado='disponible'):
        super().__init__(titulo, estado)
        self.edicion = edicion
        self.periodicidad = periodicidad

class MaterialDigital(Material):
    def __init__(self, titulo, tipo_archivo, enlace_descarga):
        super().__init__(titulo, estado='disponible')
        self.tipo_archivo = tipo_archivo
        self.enlace_descarga = enlace_descarga

class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Usuario(Persona):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.materiales_prestados = []
        self.multa = 0

    def consultar_catalogo(self, catalogo):
        return catalogo.listar_materiales()

class Bibliotecario(Persona):
    def agregar_material(self, sucursal, material):
        sucursal.catalogo.append(material)

    def gestionar_prestamo(self, usuario, material, sucursal):
        if material in sucursal.catalogo and material.estado == 'disponible':
            material.estado = 'prestado'
            usuario.materiales_prestados.append(material)
            sucursal.catalogo.remove(material)
            return Prestamo(usuario, material)
        return None

    def transferir_material(self, material, sucursal_origen, sucursal_destino):
        if material in sucursal_origen.catalogo:
            sucursal_origen.catalogo.remove(material)
            sucursal_destino.catalogo.append(material)

class Sucursal:
    def __init__(self, nombre):
        self.nombre = nombre
        self.catalogo = []

    def buscar_material(self, titulo):
        return next((m for m in self.catalogo if m.titulo == titulo), None)

class Prestamo:
    def __init__(self, usuario, material):
        self.usuario = usuario
        self.material = material
        self.fecha_prestamo = datetime.now()
        self.fecha_devolucion = self.fecha_prestamo + timedelta(days=14)

    def devolver_material(self, sucursal):
        if datetime.now() > self.fecha_devolucion:
            dias_retraso = (datetime.now() - self.fecha_devolucion).days
            Penalizacion.aplicar_multa(self.usuario, dias_retraso)
        self.material.estado = 'disponible'
        sucursal.catalogo.append(self.material)
        self.usuario.materiales_prestados.remove(self.material)

class Penalizacion:
    @staticmethod
    def aplicar_multa(usuario, dias_retraso):
        usuario.multa += dias_retraso * 2

class Catalogo:
    @staticmethod
    def listar_materiales(sucursales):
        materiales_disponibles = []
        for sucursal in sucursales:
            materiales_disponibles.extend([m.titulo for m in sucursal.catalogo if m.estado == 'disponible'])
        return materiales_disponibles

sucursal1 = Sucursal("Biblioteca Central")
sucursal2 = Sucursal("Biblioteca Sur")

libro1 = Libro("1984", "George Orwell", "Distop√≠a")
libro2 = Libro("El Principito", "Antoine de Saint-Exup√©ry", "Infantil")
revista1 = Revista("National Geographic", "Edici√≥n 2023", "Mensual")
digital1 = MaterialDigital("Curso Python", "PDF", "https://cursoselpython.com/curso-python")

sucursal1.catalogo.extend([libro1, revista1])
sucursal2.catalogo.append(libro2)

usuario1 = Usuario("Juan Jorobas")
bibliotecario1 = Bibliotecario("Ana Lisa")

prestamo1 = bibliotecario1.gestionar_prestamo(usuario1, libro1, sucursal1)

print("Materiales disponibles:", Catalogo.listar_materiales([sucursal1, sucursal2]))
print("Pr√©stamos activos:")
if prestamo1:
    print(f"{prestamo1.usuario.nombre} ha prestado '{prestamo1.material.titulo}' hasta {prestamo1.fecha_devolucion}")
print(f"Multa de {usuario1.nombre}: {usuario1.multa}")
```
